```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
data_car = pd.read_csv("USA_cars_datasets.csv")
del data_car['Unnamed: 0']
df = data_car['brand'].value_counts().rename_axis('cars_name').reset_index(name='counts').head(10)
plt.barh(df['cars_name'], df['counts'])
plt.xlabel('Quantity')
plt.ylabel('Car names')
plt.title('Top of american cars')
plt.grid()
```

на основании считанного датафрейма можно сказать, что  ford - самая популярная марка американских машин

```python

dataset = data_car.groupby('color').count().sort_values('brand',ascending = False).head(10)
dataset.plot(kind='pie', y='brand')
```

Как видно из этой диаграммы, самый популярный цвет у машин - белый. Затем идет черный.

```python
df = data_car.groupby('brand').mean().sort_values(by='price', ascending=False).head(10)
df[['price']].plot()
plt.xlabel('car name')
plt.ylabel('price')
plt.grid()
```
